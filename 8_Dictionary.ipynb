Dictionary

1. Dictionary --> {key : value} --> {'key' : value}
2. It used to store data in key value pair
3. Dictionary is mutable (key are immutable, and values are mutable)
4. Duplicate keys are not allowed
5. Duplicate values are allowed
6. comma seperated


````python
dict1 = {'name':'Ajay'}

print(dict1)
print(len(dict1))
````

````python
dict1 = {'name':'Ajay','age':25}

print(dict1)
print(len(dict1))
````

````python
dict1 = {'name':'Ajay','age':25,'city' : 'Pune', 'State' : 'Maharashtra'}

print(dict1)
print(len(dict1))
````

Access Dictionary Items

````python
access_dict = {'name':'Ajay','age':25,'city' : 'Pune', 'State' : 'Maharashtra'}
````

````python
access_dict['age']
````

````python
access_dict['name']
````

````python
# it gives output with keys only
for i in access_dict:
  print(i)
````

````python
# .keys() ---> to access keys
for i in access_dict.keys():
  print(i)
````

````python
#.values() ----> to access values
for i in access_dict.values():
  print(i)
````

````python
# .items() ----> to access key, value pairs
for i,v in access_dict.items():
  print(i,v)
````

````python
# .items() ----> to access key, value pairs
for i,v in access_dict.items():
  print(i,":",v)
````

For loop in dict

````python
#zip with set ---> select random mapping
subject = {'Phy','Che','Math','Eng'}
marks = {95,80,100,98}
# {'Phy':95,'Che':80....}
sub_marks = dict(zip(subject,marks))
print(sub_marks)
````

````python
#zip with list ---> accurate mapping
subject = ['Phy','Che','Math','Eng']
marks = [95,80,100,98]
# {'Phy':95,'Che':80....}
sub_marks = dict(zip(subject,marks))
print(sub_marks)
````

````python
subject = ['Phy','Che','Math','Eng']
marks = [95,80,100,98]

result = {}
for i,j in enumerate(subject): #0
  result[j] = marks[i]
print(result)
````

````python
string_dict = "Todays Date is 9th sept and Todays weather is humid and"

result = {} # 'todays':2
string_dict = string_dict.lower()
li = string_dict.split()
print(li)
for word in li:
  result[word] = string_dict.count(word)
print()
print(result)
````

````python
#ASCII values

upper_case = {}

for i in range(65,91):
  upper_case[chr(i)] = i

print(upper_case)
````

````python
#ASCII values

lower_case = {}

for i in range(97,123):
  lower_case[chr(i)] = i

print(lower_case)
````
