Assigning single value to multiple variables

```python
a = 100
b= 100
c = 100

print(f'I am printing {a},{b},{c}')


# print("I am printing a",a)
```

```python
a = b = c = 100
print(f'I am printing {a},{b},{c}')
```

```python
a = b = c = 'python'
print(f'I am printing {a},{b},{c}')
```

```python
a = 100; b = 101; c = 103
print(f'I am printing {a},{b},{c}')
```
