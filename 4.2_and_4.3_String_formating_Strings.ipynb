String Formatting in Python
1. foramtiing with format() method
2. formating with string literals, called f-string
3. formating with % operator

```python
#1. format() ---> follow the sequence

a = 100
b = 200

# print("addition of {} and {} is {}")
print("addition of {} and {} is {}".format(a,b,a+b))
# print("addition of {} and {} is {}".format(a,b,a*b))
```

```python
#1. format()

# a = 100
# b = 200
name1 = 'Deepak'
name2 = 'Avinash'

print("{} and {} are best friends".format(name1,name2))
# print("addition of {} and {} is {}")
# print("addition of {} and {} is {}".format(a,b,a+b))
# print("addition of {} and {} is {}".format(a,b,a*b))
```

```python
a = 100
b = 200

print("multiplication of {} and {} is {}".format(a,b,a*b))
```

```python
#2. f-string
a = 100
b = 20

print(f"addition of {b} and {a} is {a+b}")
```

```python
base_url = 'hdbsludcb'
prefix = 'kjanc'
suffix = 'dk'

url = base_url + prefix + suffix

print(url)
```

```python
base_url = 'https://www.canva.com/'
project_url = 'projects'
temp_ur = 'templates'

template = base_url + temp_ur
project = base_url + project_url

print(f'url to access project is :{project} and url for templates is: {template}')
```

```python
base_url = 'https://www.canva.com/'

li = ['projects','templates','ai','brand']
for i in li:
    print(f'url to access project is :{base_url}{i}')
```

```python
#3. Formatting with % Operator
# %d ---> int
# %f ---> float
# %s ---> str

#Example:

a = 10.2
b = 20.5
c = a+b

str_operator = "addition of %f and %f is " %(a,b,c)
str_operator
```

```python

a = 10.2
b = 20.5
c = a+b

str_operator = "addition of %f and %f is %f" %(a,b,c)
str_operator
```
