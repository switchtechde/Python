What are functions in Python?
A function in python is ablock of reusable code that  performs a specific task

WHy use Function?
1. Reusability : write once and we can use it multiple times
2. Modularity : Breaks a big problem into small
3. Readability : make code cleaner

Components of Function:
1. keyword (def) --> used to define a function
2. Function name --> identifier of function
3. Parameters (optional)--> input variables
4. Docstring (Optional) --> Documentation of what function does
5. Body --> block of code
6. Return statement (Optional)seds value back to caller

*args vs **kwargs

*args : postional arguments or values without variable name
**kwargs : keyword argument. It means its output comes in key-value pair that is dictionary

```python
# sum()

def add_all(*args):
  return sum(args)

print(add_all(1,2,3))

print(add_all(10,20,50,5,44,66))
```

```python
def print_details(**kwargs):
  for key,value in kwargs.items():
    print(f'{key}-->{value}')

print_details(name = 'Shubham',age = 40, city = 'Pune')
```

```python
import  time #timne

def time_tracker(func): #decorator
  def wrapper(*agrs, **kwargs): #wrapper
    start = time.time() #start timer
    print(f"start time : {start}")

    result = func(*agrs, **kwargs)

    end = time.time()
    print(f"end time : {end}")
    print(f'Dursation : {end - start} seconds')
    return result
  return wrapper
```

```python
@time_tracker
def slow_function():
  # time.sleep(5)
  # print(time.time())
  print("sleep time over")

slow_function()
```
