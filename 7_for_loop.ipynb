```python
string = 'same fuctionality but diffrent syntax and condition checking'
length = len(string)
# print(length)

for index in range(length): #0
  print(string[index]) #string[0]

#round_1 = index = 0, string[0] --> s
#round_2 = index = 1, string[1] --> a
#round_3 = index = 2, string[2] --> m

```

```python
#5. Enumerate ---> output with index and its value

fruit_list_enum  = ["apple","grapes","cherry","banana"]
 #                       0       1       2       3
for ind,value in enumerate(fruit_list_enum):
  print(ind,value)
```

```python
#5. Enumerate ---> output with index and its value

fruit_list_enum  = ["apple","grapes","cherry","banana"]
 #                       0       1       2       3
for i,j in enumerate(fruit_list_enum):
  print(i,j)
```

```python
#6. zip
name = ["shubham",'Nisha','Keshava','Raghav',"ajay"]
age = [5,6,7,8,9]

for n,a in zip(name,age):
  # print(n,a)
  print(f"age of {n} is {a}")
```
