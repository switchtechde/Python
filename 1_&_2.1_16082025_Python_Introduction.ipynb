Types of Variable

1.   Local Variable : Declare inside the fuction (def)
2.   Global variable : We can declare it outside the function


```python
message = "I am global variable" #global variable

def show_message():
  note = "local variable" #local variable
    
  print(message)
  print(note)

show_message()
print("local variable",note)
print("global variable",message)
```

```python
x = 10
def lets_check():
    global x
    x = 20
lets_check()
print(x)
```
