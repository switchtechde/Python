1. Break : immediately exits the loop (for and while), no further iteration are executed.
Use : when you want to stop once a condition met

```python
for i in range(6): #0,1,2,3,4
  print("before break",i)
  if i == 3:

    break
  print("after break",i)
```

# continue :  skip the current iteration and mov to the next iteration

usecase : to skip the specific value but dont want to discontinue the loop

```python
for i in range(6):
  print("before break",i)
  if i == 3:

    continue
  print("after break",i)
```

```python
li = [1,2,3,5,4,6,9,8]
for i in li:
  if i == 5:
    continue
  print(i)
```

# pass : if we dont want to give any code but want ot run the code we simple use pass. It means "do nothing"


```python
for i in range(6): #0,1,2,3,4
  if i == 3:
    pass
  print("after break",i)
```

# Deep copy vs shallow copy

```python
import copy

li1 = [[1,2],[3,4]]
        # 0     1
shallow = copy.copy(li1)

print("ORIGINAL-->",li1)
print("SHALLOW--->",shallow)

#LETS MODIFY
shallow[0][0] = 100

print("After modification")
print("ORIGINAL-->",li1)
print("SHALLOW--->",shallow)
```

```python
import copy as cc

li2 = [[1,2],[3,4]]
        # 0     1
deep_copy = cc.deepcopy(li2)

print("ORIGINAL-->",li2)
print("deep_copy--->",deep_copy)

#LETS MODIFY
deep_copy[0][0] = 100
print()
print("After modification")
print("ORIGINAL-->",li2)
print("deep_copy--->",deep_copy)
```
