try:
  #code that might cause an error
except SomeError: #when we know the error
  #code that runs if error happen
---------

try: # expected that it will throw error for exceptions
  print("Executes if No error")
except Exception as e: #when we dont know the the error
  print("Executes If error occur",e)
else:
  print("executes with try, not with except")
finally:
  print("always execute, no matter error occur or not")

In short, if error occur then "try + except"
if error not occur then "try + else"
