SET ---> {}

1. A set is unordered collection of unique elements
2. It is mutable
3. **Duplicates are not allowed or it contains only uniques elements


```python
set_1 = {1,2,3,4}
print(set_1)
```

```python
# Removing duplicates from list
li= [2,3,5,5,4,4,4,2]
var = set(li)
print(var)
```

Functions in set

```python
# add()
set_2 = {1,2,3,4}
set_2.add(6)
print(set_2)
```

```python
#update()
set_3= {1,2,3,4}
set_3.update([5,6,7])
print(set_3)
```

```python
# remove()
# it excepts only one argument
set_4 = {1, 2, 3, 4, 5, 6, 7}
set_4.remove(2)

print(set_4)
```

```python
set_4 = {1, 2, 3, 4, 5, 6, 7}
set_4.remove(9)

print(set_4)
```

```python
# discard() ---> to avoid error if element is absent
set_4 = {1, 2, 3, 4, 5, 6, 7}
set_4.discard(9)

print(set_4)
```

```python
# pop()
set_5 = {1, 2, 3, 4, 5, 6, 7}

set_5_pop = set_5.pop()

print(set_5)
print(set_5_pop)
```

```python
s = {1,2,3,4}
x = s.pop()

print("Removed:",x)

print("Remaining",s)
```

```python
# clear()
set_6 = {1, 2, 3, 4, 5, 6, 7}
set_6.clear()

print(set_6)
```

```python
# union() --->combine
s1 = {1,2,3}
s2 = {3,4,5}
print(s1.union(s2))
```

```python
# intersection() --> finds common elements
s1 = {1,2,3}
s2 = {3,4,5}
print(s1.intersection(s2))
```

```python
# diffrence()
s1 = {1,2,3,4}
s2 = {3,4,5}
print(s1 - s2)
```

```python
#issubset()
s1 = {1,2,3,4}
s2 = {1,2}
s3 = {3,4}
s4 = {2,4}
s5 = {1,5}
print(s2.issubset(s1))
print(s3.issubset(s1))
print(s4.issubset(s1))
print(s5.issubset(s1)) #false

```
