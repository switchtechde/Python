# loc[]
1. label based indexing
2. you selecct row/column (name)
3. include last/end value
4. # syntax --> df.loc[start:end,"start column_name":"end columns_name"]
# iloc[] : i for index
1. integer position-based indexing
2. row/column numbers(position)
3. Exclude the end value
4. # syntax --> df.loc[start row_index : end row_index,"start column_index":"end column_index"]

```python
df.loc[[0,1,2]]
```

```python
df.loc[1:3]
```

```python
df.loc[[1,3],["Name",'Country']]
```

```python
# syntax --> df.loc[start:end,"start column_name":"end columns_name"]
df.loc[1:3,"Name":"Country"]
```

```python
# syntax --> df.loc[start row_index : end row_index,"start column_index":"end column_index"]
df.iloc[1:5,1:5]
```
